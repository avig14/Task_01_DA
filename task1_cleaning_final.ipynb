```
import pandas as pd

# Load dataset
df = pd.read_csv("KaggleV2-May-2016.csv")

# Inspect
print(df.head())
print(df.info())
print(df.isnull().sum())

# Remove duplicates
df = df.drop_duplicates()

# Standardize column names
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(" ", "_")
              .str.replace("-", "_")
)

# Convert date columns
df['scheduledday'] = pd.to_datetime(df['scheduledday'], errors='coerce')
df['appointmentday'] = pd.to_datetime(df['appointmentday'], errors='coerce')

# Standardize text columns
df['gender'] = df['gender'].astype(str).str.upper()
df['neighbourhood'] = df['neighbourhood'].astype(str).str.title()

# Convert no-show column
if 'no_show' in df.columns:
    df['no_show'] = df['no_show'].map({'Yes': 1, 'No': 0})
elif 'no-show' in df.columns:
    df['no_show'] = df['no-show'].map({'Yes': 1, 'No': 0})
    df = df.drop(columns=['no-show'])

# Fix age
df['age'] = pd.to_numeric(df['age'], errors='coerce').fillna(0).astype(int)
df['age'] = df['age'].clip(lower=0)

# Save cleaned dataset
df.to_csv("cleaned_medical_appointments.csv", index=False)

df.head()
```
